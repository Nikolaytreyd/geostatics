# Типы распределений

1. Равномерное распределение (Uniform Distribution):
   - Особенности: Все значения в заданном интервале имеют одинаковую вероятность.
   - Применение: Используется для моделирования случайных событий, когда все значения равновероятны, таких как выбор случайного числа из заданного диапазона.
   - Функция NumPy: `np.random.uniform()`

2. Распределение целых чисел (Integer Distribution):
   - Особенности: Генерирует случайные целые числа в заданном диапазоне.
   - Применение: Часто используется для моделирования дискретных случайных событий, таких как бросок монеты или подбрасывание кубика.
   - Функция NumPy: `np.random.randint()`

3. Нормальное распределение (Normal Distribution):
   - Особенности: Имеет колоколообразную форму с симметричной кривой вокруг среднего значения.
   - Применение: Широко используется в статистике и анализе данных для моделирования случайных величин, таких как рост людей или ошибка измерений.
   - Функция NumPy: `np.random.normal()`

4. Экспоненциальное распределение (Exponential Distribution):
   - Особенности: Имеет экспоненциальную форму и используется для моделирования времени между последовательными событиями.
   - Применение: Часто используется в теории надежности, финансовой математике и других областях, где необходимо моделировать время до наступления события.
   - Функция NumPy: `np.random.exponential()`

5. Распределение Пуассона (Poisson Distribution):
   - Особенности: Используется для моделирования количества событий, происходящих в заданный промежуток времени или в заданной области пространства.
   - Применение: Часто используется в теории массового обслуживания, телекоммуникациях и других областях, где необходимо моделировать случайные счетчики событий.
   - Функция NumPy: `np.random.poisson()`

6. Гамма-распределение (Gamma Distribution):
   - Особенности: Используется для моделирования времени наступления события, когда время до наступления события имеет форму, близкую к экспоненциальному распределению.
   - Применение: Часто используется в физике, биологии и других областях, где время до наступления события может быть случайным и иметь широкий диапазон значений.
   - Функция NumPy: `np.random.gamma()`

7. Бета-распределение (Beta Distribution):
   - Особенности: Используется для моделирования случайных величин, ограниченных между 0 и 1.
   - Применение: Часто используется в байесовской статистике, теории очередей и других областях, где необходимо моделировать случайные вероятности или доли.
   - Функция NumPy: `np.random.beta()`

8. Биномиальное распределение (Binomial Distribution):
   - Особенности: Используется для моделирования бинарных случайных величин, таких как количество успехов в серии независимых испытаний.
   - Применение: Часто используется в теории вероятностей, статистике и других областях, где необходимо моделировать количество успехов или неудач в серии испытаний.
   - Функция NumPy: `np.random.binomial()`

# Статистические метрики

1. **Histogram** *(гистограмма)*: Этот график позволяет визуализировать частоту появления значений в определенных интервалах и может помочь определить форму распределения.

- Описание: Гистограмма представляет собой графическое представление распределения данных, позволяющее визуализировать частоту появления значений в определенных интервалах.
- Применение: Гистограмма может помочь определить форму распределения данных, выявить выбросы или аномалии, а также сравнить распределения разных переменных или групп данных.

2. **Frequency distribution curve** *(линейный график)*: График позволяет наглядно представить, как вероятность или частота различных значений случайной величины распределены по ее возможным значениям. Это может помочь в понимании формы распределения данных и выявлении характеристик, таких как среднее значение, медиана и стандартное отклонение.

- Описание: Этот график представляет собой линейный график, который показывает, как вероятность или частота различных значений случайной величины распределены по ее возможным значениям.
- Применение: График частотного распределения используется для визуализации формы распределения данных и выявления характеристик, таких как среднее значение, медиана и стандартное отклонение. Он может помочь в анализе и сравнении различных распределений данных.

3. **Scatter plot** *(график рассеяния)*: Этот график используется для визуализации двух переменных и может помочь определить наличие какой-либо структуры или зависимости в данных.

- Описание: График рассеяния используется для визуализации двух переменных и показывает точки, представляющие значения этих переменных на плоскости. Каждая точка представляет собой одно наблюдение.
- Применение: График рассеяния может помочь определить наличие какой-либо структуры или зависимости между двумя переменными. Он может использоваться для поиска корреляции, выявления выбросов или группировки данных.

4. **Violinplot** *(скрипичная диаграмма)*: Этот график позволяет визуально оценить основные статистические характеристики распределения, такие как медиана, квартили и выбросы.Текст, выделенный полужирным шрифтом.

- Описание: Скрипичная диаграмма представляет собой комбинацию гистограммы и графика плотности. Она позволяет визуально оценить основные статистические характеристики распределения, такие как медиана, квартили и выбросы.
- Применение: Скрипичная диаграмма часто используется для сравнения распределений разных групп данных или для анализа одного распределения с различными условиями или факторами.

5. **Kernel density plot** *(график плотности ядра)*: Этот график представляет собой сглаженную оценку плотности распределения данных и может помочь определить форму и гладкость распределения.

- Описание: График плотности ядра представляет собой сглаженную оценку плотности распределения данных. Он помогает определить форму и гладкость распределения.
- Применение: График плотности ядра используется для визуализации распределений данных и сравнения формы распределений разных переменных или групп данных.

6. **Empirical cumulative distribution function (ECDF)** *(линейный график)*: Этот график показывает, какую долю данных составляют значения меньше или равные определенному пороговому значению. Он может помочь визуализировать форму и хвосты распределения.

- Описание: График представляет собой линию, которая начинается с нуля и постепенно увеличивается до единицы по мере увеличения порогового значения. График ECDF помогает визуализировать форму распределения данных, а также оценить вероятности и квантили. Он также может использоваться для сравнения нескольких распределений данных и анализа их хвостовых областей.
- Применение: График ECDF часто используется для анализа распределений данных, поиска квантилей, оценки вероятностей и сравнения нескольких распределений данных.

#  Непрерывное распределение

```python
def plot_uniform_distribution(width):
    data = np.random.uniform(0, width+1, 100_000)

    plt.style.use('seaborn-whitegrid')
    sns.set_style("dark")
    fig, axs = plt.subplots(3, 2, figsize=(12, 10))

    fontsize_label = 6
    fontsize_title = 7
    fontsize_tick = 6
    name = 'Continuous'

    # Гистограмма распределения
    axs[0, 0].hist(data, bins=10)
    axs[0, 0].set_xlabel('Value', fontsize=fontsize_label)
    axs[0, 0].set_ylabel('Frequency', fontsize=fontsize_label)
    axs[0, 0].set_title(f'{name} distribution', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[0, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Кривая распределения частот
    bins = 10
    hist, bin_edges = np.histogram(data, bins=bins)
    freq_dist = hist / len(data)
    axs[0, 1].plot(bin_edges[:-1], freq_dist, linewidth=4)
    axs[0, 1].set_xlabel('Value', fontsize=fontsize_label)
    axs[0, 1].set_ylabel('Frequency', fontsize=fontsize_label)
    axs[0, 1].set_title('Frequency distribution curve', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[0, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Точечый график
    x = np.random.uniform(0, width+1, 1000)
    y = np.random.uniform(0, width+1, 1000)
    axs[1, 1].scatter(x, y)
    axs[1, 1].set_xlabel('X', fontsize=fontsize_label)
    axs[1, 1].set_ylabel('Y', fontsize=fontsize_label)
    axs[1, 1].set_title('Scatter plot', fontsize=fontsize_title,loc='left', fontweight='700')
    axs[1, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Японская свеча
    axs[1, 0].violinplot(dataset=[data, x, y], showmeans=False, showmedians=True)
    axs[1, 0].set_ylabel('Value', fontsize=fontsize_label)
    axs[1, 0].set_title('Violinplot', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[1, 0].set_xticks([y + 1 for y in range(3)], labels=['Data', 'X', 'Y'])
    axs[1, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Kernel density plot
    sns.kdeplot(data, ax=axs[2, 0], linewidth=4)
    axs[2, 0].set_xlabel('Value', fontsize=fontsize_label)
    axs[2, 0].set_ylabel('Density', fontsize=fontsize_label)
    axs[2, 0].set_title('Kernel density plot', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[2, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Empirical cumulative distribution function (ECDF) plot
    ecdf = ECDF(data)
    axs[2, 1].plot(ecdf.x, ecdf.y, linewidth=4)
    axs[2, 1].set_xlabel('Value', fontsize=fontsize_label)
    axs[2, 1].set_ylabel('Cumulative Probability', fontsize=fontsize_label)
    axs[2, 1].set_title('ECDF plot', fontsize=fontsize_title,loc='left', fontweight='700')
    axs[2, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Статические метрики для data
    mode_data = np.round(np.mean(data), 2)
    median_data = np.round(np.median(data), 2)
    std_dev_data = np.round(np.std(data), 2)
    mean_val_data = np.round(np.mean(data), 2)
    percentile_10_data = np.round(np.percentile(data, 10), 2)
    percentile_90_data = np.round(np.percentile(data, 90), 2)

    # Статические метрики для x
    mode_x = np.round(np.mean(x), 2)
    median_x = np.round(np.median(x), 2)
    std_dev_x = np.round(np.std(x), 2)
    mean_val_x = np.round(np.mean(x), 2)
    percentile_10_x = np.round(np.percentile(x, 10), 2)
    percentile_90_x = np.round(np.percentile(x, 90), 2)

    # Статические метрики для y
    mode_y = np.round(np.mean(y), 2)
    median_y = np.round(np.median(y), 2)
    std_dev_y = np.round(np.std(y), 2)
    mean_val_y = np.round(np.mean(y), 2)
    percentile_10_y = np.round(np.percentile(y, 10), 2)
    percentile_90_y = np.round(np.percentile(y, 90), 2)

    text = f"{name} distribution data\nMode: {mode_data}\nMedian: {median_data}\nStandard Deviation: {std_dev_data}\nMean: {mean_val_data}\n10th Percentile: {percentile_10_data}\n90th Percentile: {percentile_90_data} \n\n\n\n" \
           f"{name} distribution x\nMode: {mode_x}\nMedian: {median_x}\nStandard Deviation: {std_dev_x}\nMean: {mean_val_x}\n10th Percentile: {percentile_10_x}\n90th Percentile: {percentile_90_x} \n\n\n\n" \
           f"{name} distribution y\nMode: {mode_y}\nMedian: {median_y}\nStandard Deviation: {std_dev_y}\nMean: {mean_val_y}\n10th Percentile: {percentile_10_y}\n90th Percentile: {percentile_90_y}"

    plt.text(1.05, 0.75, text, transform=plt.gca().transAxes)

    plt.tick_params(axis='both', which='major', labelsize=5)
    plt.show()

interact(plot_uniform_distribution, width=FloatSlider(min=0, max=100, step=1, value=50))
```

```python
def plot_normal_distribution(mean, sqr):
    data = np.random.normal(mean, sqr, 100_000)
    plt.style.use('seaborn-whitegrid')
    sns.set_style("dark")
    fig, axs = plt.subplots(3, 2, figsize=(12, 10))

    fontsize_label = 6
    fontsize_title = 7
    fontsize_tick = 6
    name = 'Normal'

    # Гистограмма распределения
    axs[0, 0].hist(data, bins=10)
    axs[0, 0].set_xlabel('Value', fontsize=fontsize_label)
    axs[0, 0].set_ylabel('Frequency', fontsize=fontsize_label)
    axs[0, 0].set_title(f'{name} distribution', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[0, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Кривая распределения частот
    bins = 10
    hist, bin_edges = np.histogram(data, bins=bins)
    freq_dist = hist / len(data)
    axs[0, 1].plot(bin_edges[:-1], freq_dist, linewidth=4)
    axs[0, 1].set_xlabel('Value', fontsize=fontsize_label)
    axs[0, 1].set_ylabel('Frequency', fontsize=fontsize_label)
    axs[0, 1].set_title('Frequency distribution curve', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[0, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Точечый график
    x = np.random.normal(mean, sqr, 1_000)
    y = np.random.normal(mean, sqr, 1_000)
    axs[1, 1].scatter(x, y)
    axs[1, 1].set_xlabel('X', fontsize=fontsize_label)
    axs[1, 1].set_ylabel('Y', fontsize=fontsize_label)
    axs[1, 1].set_title('Scatter plot', fontsize=fontsize_title,loc='left', fontweight='700')
    axs[1, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Японская свеча
    axs[1, 0].violinplot(dataset=[data, x, y], showmeans=False, showmedians=True)
    axs[1, 0].set_ylabel('Value', fontsize=fontsize_label)
    axs[1, 0].set_title('Violinplot', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[1, 0].set_xticks([y + 1 for y in range(3)], labels=['Data', 'X', 'Y'])
    axs[1, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Kernel density plot
    sns.kdeplot(data, ax=axs[2, 0], linewidth=4)
    axs[2, 0].set_xlabel('Value', fontsize=fontsize_label)
    axs[2, 0].set_ylabel('Density', fontsize=fontsize_label)
    axs[2, 0].set_title('Kernel density plot', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[2, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Empirical cumulative distribution function (ECDF) plot
    ecdf = ECDF(data)
    axs[2, 1].plot(ecdf.x, ecdf.y, linewidth=4)
    axs[2, 1].set_xlabel('Value', fontsize=fontsize_label)
    axs[2, 1].set_ylabel('Cumulative Probability', fontsize=fontsize_label)
    axs[2, 1].set_title('ECDF plot', fontsize=fontsize_title,loc='left', fontweight='700')
    axs[2, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Статические метрики для data
    mode_data = np.round(np.mean(data), 2)
    median_data = np.round(np.median(data), 2)
    std_dev_data = np.round(np.std(data), 2)
    mean_val_data = np.round(np.mean(data), 2)
    percentile_10_data = np.round(np.percentile(data, 10), 2)
    percentile_90_data = np.round(np.percentile(data, 90), 2)

    # Статические метрики для x
    mode_x = np.round(np.mean(x), 2)
    median_x = np.round(np.median(x), 2)
    std_dev_x = np.round(np.std(x), 2)
    mean_val_x = np.round(np.mean(x), 2)
    percentile_10_x = np.round(np.percentile(x, 10), 2)
    percentile_90_x = np.round(np.percentile(x, 90), 2)

    # Статические метрики для y
    mode_y = np.round(np.mean(y), 2)
    median_y = np.round(np.median(y), 2)
    std_dev_y = np.round(np.std(y), 2)
    mean_val_y = np.round(np.mean(y), 2)
    percentile_10_y = np.round(np.percentile(y, 10), 2)
    percentile_90_y = np.round(np.percentile(y, 90), 2)

    text = f"{name} distribution data\nMode: {mode_data}\nMedian: {median_data}\nStandard Deviation: {std_dev_data}\nMean: {mean_val_data}\n10th Percentile: {percentile_10_data}\n90th Percentile: {percentile_90_data} \n\n\n\n" \
           f"{name} distribution x\nMode: {mode_x}\nMedian: {median_x}\nStandard Deviation: {std_dev_x}\nMean: {mean_val_x}\n10th Percentile: {percentile_10_x}\n90th Percentile: {percentile_90_x} \n\n\n\n" \
           f"{name} distribution y\nMode: {mode_y}\nMedian: {median_y}\nStandard Deviation: {std_dev_y}\nMean: {mean_val_y}\n10th Percentile: {percentile_10_y}\n90th Percentile: {percentile_90_y}"

    plt.text(1.05, 0.75, text, transform=plt.gca().transAxes)

    plt.tick_params(axis='both', which='major', labelsize=5)
    plt.show()

interact(plot_normal_distribution,
         mean=FloatSlider(min=0, max=50, step=0.1, value=25),
         sqr=FloatSlider(min=0.1, max=10, step=0.1, value=5),
         )
```

```python
def plot_beta_distribution(a, b):
    data = np.random.beta(a, b, 100_000)

    plt.style.use('seaborn-whitegrid')
    sns.set_style("dark")
    fig, axs = plt.subplots(3, 2, figsize=(12, 10))

    fontsize_label = 6
    fontsize_title = 7
    fontsize_tick = 6
    name = 'Beta'

    # Гистограмма распределения
    axs[0, 0].hist(data, bins=10)
    axs[0, 0].set_xlabel('Value', fontsize=fontsize_label)
    axs[0, 0].set_ylabel('Frequency', fontsize=fontsize_label)
    axs[0, 0].set_title(f'{name} distribution', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[0, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Кривая распределения частот
    bins = 10
    hist, bin_edges = np.histogram(data, bins=bins)
    freq_dist = hist / len(data)
    axs[0, 1].plot(bin_edges[:-1], freq_dist, linewidth=4)
    axs[0, 1].set_xlabel('Value', fontsize=fontsize_label)
    axs[0, 1].set_ylabel('Frequency', fontsize=fontsize_label)
    axs[0, 1].set_title('Frequency distribution curve', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[0, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Точечый график
    x = np.random.beta(a, b, 1_000)
    y = np.random.beta(a, b, 1_000)
    axs[1, 1].scatter(x, y)
    axs[1, 1].set_xlabel('X', fontsize=fontsize_label)
    axs[1, 1].set_ylabel('Y', fontsize=fontsize_label)
    axs[1, 1].set_title('Scatter plot', fontsize=fontsize_title,loc='left', fontweight='700')
    axs[1, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Японская свеча
    axs[1, 0].violinplot(dataset=[data, x, y], showmeans=False, showmedians=True)
    axs[1, 0].set_ylabel('Value', fontsize=fontsize_label)
    axs[1, 0].set_title('Violinplot', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[1, 0].set_xticks([y + 1 for y in range(3)], labels=['Data', 'X', 'Y'])
    axs[1, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Kernel density plot
    sns.kdeplot(data, ax=axs[2, 0], linewidth=4)
    axs[2, 0].set_xlabel('Value', fontsize=fontsize_label)
    axs[2, 0].set_ylabel('Density', fontsize=fontsize_label)
    axs[2, 0].set_title('Kernel density plot', fontsize=fontsize_title, loc='left', fontweight='700')
    axs[2, 0].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Empirical cumulative distribution function (ECDF) plot
    ecdf = ECDF(data)
    axs[2, 1].plot(ecdf.x, ecdf.y, linewidth=4)
    axs[2, 1].set_xlabel('Value', fontsize=fontsize_label)
    axs[2, 1].set_ylabel('Cumulative Probability', fontsize=fontsize_label)
    axs[2, 1].set_title('ECDF plot', fontsize=fontsize_title,loc='left', fontweight='700')
    axs[2, 1].tick_params(axis='both', which='major', labelsize=fontsize_tick)

    # Статические метрики для data
    mode_data = np.round(np.mean(data), 2)
    median_data = np.round(np.median(data), 2)
    std_dev_data = np.round(np.std(data), 2)
    mean_val_data = np.round(np.mean(data), 2)
    percentile_10_data = np.round(np.percentile(data, 10), 2)
    percentile_90_data = np.round(np.percentile(data, 90), 2)

    # Статические метрики для x
    mode_x = np.round(np.mean(x), 2)
    median_x = np.round(np.median(x), 2)
    std_dev_x = np.round(np.std(x), 2)
    mean_val_x = np.round(np.mean(x), 2)
    percentile_10_x = np.round(np.percentile(x, 10), 2)
    percentile_90_x = np.round(np.percentile(x, 90), 2)

    # Статические метрики для y
    mode_y = np.round(np.mean(y), 2)
    median_y = np.round(np.median(y), 2)
    std_dev_y = np.round(np.std(y), 2)
    mean_val_y = np.round(np.mean(y), 2)
    percentile_10_y = np.round(np.percentile(y, 10), 2)
    percentile_90_y = np.round(np.percentile(y, 90), 2)

    text = f"{name} distribution data\nMode: {mode_data}\nMedian: {median_data}\nStandard Deviation: {std_dev_data}\nMean: {mean_val_data}\n10th Percentile: {percentile_10_data}\n90th Percentile: {percentile_90_data} \n\n\n\n" \
           f"{name} distribution x\nMode: {mode_x}\nMedian: {median_x}\nStandard Deviation: {std_dev_x}\nMean: {mean_val_x}\n10th Percentile: {percentile_10_x}\n90th Percentile: {percentile_90_x} \n\n\n\n" \
           f"{name} distribution y\nMode: {mode_y}\nMedian: {median_y}\nStandard Deviation: {std_dev_y}\nMean: {mean_val_y}\n10th Percentile: {percentile_10_y}\n90th Percentile: {percentile_90_y}"

    plt.text(1.05, 0.75, text, transform=plt.gca().transAxes)

    plt.tick_params(axis='both', which='major', labelsize=5)
    plt.show()


interact(plot_beta_distribution,
         a=FloatSlider(min=0.01, max=10, step=0.01, value=5),
         b=FloatSlider(min=0.01, max=10, step=0.01, value=5),
         )
```
